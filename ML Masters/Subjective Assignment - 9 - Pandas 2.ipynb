Q25. Write a pandas program to make partition each of the passengers
into 4 categories based on their age.
 
Note: Age categories- (0, 10), (10, 30), (30, 60), (60, 80)

```python
def categorize_age(age):
    
    result = ' '
    if age >= 0 and age < 10:
        result = 'Child'
    elif age >= 10 and age < 30:
        result = 'Youth'
    elif age >= 30 and age < 60:
        result = 'Mid-Aged'
    elif age >= 60:
        result = 'Old'
    return result


df['age_category'] = df['age'].apply(categorize_age)

df
```

## Q38. Write a pandas program to create a comparison of the top 10 years in which the (UFO) was sighted VS each Month.

```python
top_sightings_years = df['year'].value_counts().reset_index()[0:10]
top_sightings_years = list(top_sightings_years['index'])
top_sightings_years
```

```python
df_top_sightings_years = df.filter(items = ['year', 'month'])
df_top_sightings_years
```

```python
def is_top_sighting_year(year):
    if year in top_sightings_years:
        return True
    else:
        return False
    
df_top_sightings_years['is_top_sight_year'] = df_top_sightings_years['year'].apply(is_top_sighting_year)
df_top_sightings_years
```

```python
df_top_sightings_years = df_top_sightings_years[df_top_sightings_years['is_top_sight_year'] == True ]
```

```python
data = pd.pivot_table(df_top_sightings_years, index = ['year'], columns = ['month'], aggfunc = 'count')
data
```
