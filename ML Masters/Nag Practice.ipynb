```python
def sortArrays(arr): 

    # Finding the length of array 'arr' 
    length = len(arr) 

    # Sorting using a single loop 
    j = 0

    while j < length - 1:               
        if arr[j] < 0:
            ArrJ = arr[j] * -1
        else:
            ArrJ = arr[j]
        
        if arr[j+1] < 0:
            ArrJ1 = arr[j+1] * -1
        else:
            ArrJ1 = arr[j+1]
        
        
        # Checking the condition for two 
        # simultaneous elements of the array 
        if (ArrJ > ArrJ1): 

            # Swapping the elements. 
            temp = arr[j] 
            arr[j] = arr[j + 1] 
            arr[j + 1] = temp 
            
            # updating the value of j = -1 
            # so after getting updated for j++ 
            # in the loop it becomes 0 and 
            # the loop begins from the start. 
            j = -1
        j += 1
    return arr 

ListB = [-8,-5,-3,-1,3,6,9]
print(sortArrays(ListB))
```

```python
# Python3 program to reverse a string

# Function to reverse each word in the string
def reverse_word(s, start, end):
	while start < end:
		s[start], s[end] = s[end], s[start]
		start = start + 1
		end -= 1


s = 'I Love Programming'

# Convert string to list to use it as a char array
s = list(s)
start = 0
while True:
	
	# We use a try catch block because for
	# the last word the list.index() function
	# returns a ValueError as it cannot find 
	# a space in the list
	try:
		# Find the next space
		end = s.index(' ', start)

		# Call reverse_word function
		# to reverse each word
		reverse_word(s, start, end - 1)

		#Update start variable
		start = end + 1

	except ValueError:

		# Reverse the last word
		reverse_word(s, start, len(s) - 1)
		break

# Reverse the entire list
s.reverse()

# Convert the list back to
# string using string.join() function
s = "".join(s)

print(s)
```

```python
def count_StringChars(InputString):
    List   = list(InputString)
    Output = []
   
    for i in List:
        
        if i not in Output:
            Output.append(i)
            Index = Output.index(i)
            Output.insert(Index+1,'1')
        
        elif i in Output:
            Index = Output.index(i)
            Output[Index+1] = int(Output[Index+1]) + 1
            Output[Index+1] = str(Output[Index+1])
        
    return "".join(Output)
```

### Case Variable Name Change

```python
def variable_conv(cls,InputString):
    
    if cls == 'Java':
        Output = ''
        List = []
        
        List = InputString.split('_')
        for i in range(1, len(List)):        
            wordI = List[i][0].upper()
            wordi = List[i][1:]          
            List[i] = wordI + wordi
        Output = "".join(List)
        
    if cls == 'C++':
        temp = InputString.lower()
        temp = list(temp)
        InputString = list(InputString)
        Output = temp.copy()
        k=0
        
        for i in range(len(InputString)):
            if InputString[i] != temp[i]:
                #print('InputString[i] = {} temp[i] = {} Output ={}'.format(InputString[i], temp[i], Output))
                Output.insert(i+k,'_')
                k=k+1
            
            
        Output = "".join(Output)       
        
    return Output    
```
